```python
import pandas as pd
import numpy as np

df = pd.read_csv('health_insurance.csv')
df.head(7)
```

Informations sur l'ensemble de données :

* **âge** - l'âge du client (en années)
* **sexe** - le sexe biologique du client (homme ou femme)
* **IMC (Indice de Masse Corporelle)** - l'indice de masse corporelle du client
* **enfants** - le nombre d'enfants que le client a
* **fumeur** - indique si le client est fumeur ou non
* **région** - la région dans laquelle le client habite
* **logcharges (logarithme des frais)** - le logarithme de la somme d'argent facturée à la compagnie d'assurance pour les soins médicaux du client (en dollars américains)

```python
df['charges'] = df['logcharges'].apply(lambda x : np.exp(x))

def status(x) :
  if (x < 18.5) : return ('Underweight')
  if (x > 18.5 and x < 24.9) : return ('Healthy Weight')
  if (x > 25 and x < 29.9) : return ('Overweight')
  else : return ('Obesity')
df['weight_status'] = df['bmi'].apply(status)

df.drop(columns='logcharges', inplace=True)
```

## Encoding

L'encodage, c'est le processus de conversion d'informations, comme du texte, des images ou des sons, en un format compréhensible par un ordinateur. Cela permet à l'ordinateur de stocker, de traiter et de transmettre ces informations. Pensez-y comme à une manière de 'coder' des données pour que l'ordinateur puisse les comprendre et les utiliser.

```python
from sklearn.preprocessing import LabelEncoder

categorial_var = ['sex', 'children', 'smoker', 'region', 'weight_status']
continuous_var = ['bmi', 'charges']

for feature in categorial_var:
    df[feature] = LabelEncoder().fit_transform(df[feature])
df.head()
```

## Normalisation

La normalisation, c'est le processus de rendre les choses plus uniformes ou standardisées. Dans le domaine de la base de données, par exemple, la normalisation signifie organiser les données d'une manière bien structurée pour éviter la redondance et garantir que chaque élément d'information a sa place appropriée. Cela facilite le stockage, la recherche et la gestion des données, ce qui les rend plus efficaces et faciles à utiliser.

Avant

```python
import matplotlib.pyplot as plt

for var in continuous_var:
    plt.figure() 
    plt.hist(df[var], bins=10, color='blue', edgecolor='black')
    plt.title('Histogramme de ' + var)
    plt.xlabel(var)
    plt.ylabel('Fréquence')
    plt.show()
```

On normalise en utilisant une fonction déja construite

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[continuous_var] = scaler.fit_transform(df[continuous_var])
df.head()
```

Après

```python
for var in continuous_var:
    plt.figure()  
    plt.hist(df[var], bins=10, color='blue', edgecolor='black')
    plt.title('Histogramme de ' + var)
    plt.xlabel(var)
    plt.ylabel('Fréquence')
    plt.show()
```

## Choix de la cible 

```python
cible = df['charges']
X = df.drop(columns='charges')
```

## Regression linéaire

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()

# Entrainement
lr.fit(X, cible)

# Prédiction
ypred = lr.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
R2 = r2_score(cible, ypred)
MSE = mean_squared_error(cible, ypred)
RMSE = np.sqrt(MSE)
print('R2 = ', R2, '\nMSE = ', MSE, '\nRMSE = ', RMSE)
```

## Apprentissage non supervisé

L'objectif de cette partie est de reconnaitre et d'associer a chaque image un numéro.

```python
from sklearn.datasets import load_digits

X, y = load_digits(return_X_y=True)

fig, ax = plt.subplots(nrows= 1, ncols = 1, figsize = (4,4))

n_img_per_row = 8
img = np.zeros((10*n_img_per_row, 10 * n_img_per_row))

for i in range(n_img_per_row) : 
    ix = 10*i + 1
    for j in range (n_img_per_row) : 
        iy = 10*j + 1
        img[ix:ix + 8, iy:iy + 8] = X[i * n_img_per_row + j].reshape((8,8))

ax.imshow(img, cmap = plt.cm.binary)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
```

```python
from sklearn.manifold import TSNE
Xemb = TSNE(n_components=2).fit_transform(X)

fig, ax = plt.subplots(nrows= 1, ncols = 1, figsize = (15,15))
xMin, xMax = np.min(Xemb, axis = 0), np.max(Xemb, axis = 0 )
Xemb = (Xemb - xMin) / (xMax - xMin)

for i in range (Xemb.shape[0]) : 
    plt.text(Xemb[i, 0], Xemb[i, 1], str(y[i]), color = plt.cm.tab10(y[i]/10.), fontdict={'weight' : 'bold', 'size' : 9})

ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
fig.tight_layout()
```
